# Code Artifacts

This file contains all codes that have been shown in the presentation and/or manuscript. This file shows how to instantiate them, and some checks that the generated schedules are correct.

## 1) BB Code
[[72, 12, 6]] - arXiv:2308.07915

```python
from macroscheduler.qecc.qalp import BB_Code
from macroscheduler.scheduling import all_schedules

code = BB_Code(
    l = 6,
    m = 6,
    A_poly = [[3, 0], [0, 1], [0, 2]],
    B_poly = [[0, 3], [1, 0], [2, 0]]
)

print(f"Code has {code.num_data_q} data qubits")

scheds = all_schedules(code)

for s_x_check, s_z_check in scheds:
    code.set_se_schedule_ir(s_x_check, s_z_check)
    code.validate_se_schedule()

print(f"Validated {len(scheds)} schedules.")
```

## 2) HGP Code
[[225, 9, 6]] code - https://doi.org/10.1103/PhysRevX.15.021065

```python
from macroscheduler.qecc.qalp import HGP, LiftedMatrix, Perm
from macroscheduler.scheduling import all_schedules

m = 3
n = 4
l = 3

A_tilde = [[[Perm.right_shift(l) ** 2] for j in range(n)] for i in range(m)]
A_tilde[0][3] = A_tilde[1][1] = []

A = LiftedMatrix(A_tilde, l).get_full_matrix()
code = HGP([A, A])

print(f"Code has {code.num_data_q} data qubits")

scheds = all_schedules(code, max_schedules=100) # they are automatically validated

print(f"Validated {len(scheds)} schedules.")
```

## 3) Quasi-cyclic LP
$B_{20}^{30}$ code - https://doi.org/10.1038/s41567-024-02479-z

```python
from macroscheduler.qecc.qalp import QALP, LiftedMatrix, Perm
from macroscheduler.scheduling import all_schedules

l = 30
A_powers = [
    [0, 0, 0, 0, 0],
    [0, 2, 14, 24, 25],
    [0, 16, 11, 14, 13]
]

m = len(A_powers)
n = len(A_powers[0])

A_tilde = [[[Perm.right_shift(l) ** A_powers[i][j]] for j in range(n)] for i in range(m)]

B_tilde = [[[Perm.right_shift(l) ** (l - A_powers[i][j])] for j in range(n)] for i in range(m)]

A = LiftedMatrix(A_tilde, l)
B = LiftedMatrix(B_tilde, l)
code = QALP([A, B])

print(f"Code has {code.num_data_q} data qubits")

scheds = all_schedules(code, max_schedules=100) # they are automatically validated

print(f"Validated {len(scheds)} schedules.")
```

## 4) Trivariate Tricycle Code
[[48, 6, 4]] - arXiv:2508.10714

```python
from macroscheduler.qecc.qalp import TT_Code, LiftedMatrix, Perm
from macroscheduler.scheduling import all_schedules

code = TT_Code(
    l=2,
    m=2,
    n=4,
    A_poly=[[0, 1, 0], [0, 0, 1], [1, 0, 1], [1, 1, 2]],
    B_poly=[[0, 1, 2], [0, 1, 3]],
    C_poly=[[0, 1, 0], [1, 1, 1]],
)

print(f"Code has {code.num_data_q} data qubits")

scheds = all_schedules(code, max_schedules=100) # they are automatically validated

print(f"Validated {len(scheds)} schedules.")
```

## 5) Kasai Code
[[9216, 4612, <48]] code - http://arxiv.org/abs/2601.08824

```python
from macroscheduler.qecc.kasai import KasaiCode, all_schedules

code = KasaiCode(
    L=12,
    J=3,
    P=768,
    f_params=[(763, 435), (679, 69), (397, 330), (61, 18), (697, 612), (373, 246)],
    g_params=[(289, 496), (257, 640), (625, 200), (41, 524), (193, 672), (449, 672)]
)

MAX_SCHEDULES = 5
scheds = all_schedules(code, max_schedules=MAX_SCHEDULES)
print(f"Found {len(scheds)} valid schedules.")
```
